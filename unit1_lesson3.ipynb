```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
%matplotlib inline
```

```python
input_data1 = pd.read_csv("teacher912.csv")
input_data2 = pd.read_csv("teacher912char.csv")
```

# Choose one variable and plot that variable four different ways.

```python

sns.boxplot(y=input_data1["average_base_pay"]).set_title("Teacher Salary Box Plot")
```

```python
plt.xlabel("High School Teacher Salaries by district")
plt.ylabel("School Districts")
plt.title("High School Teacher Salaries in Texas\n (population)")
plt.hist(input_data1.average_base_pay)
```

```python
pop_mean = np.mean(input_data1['average_base_pay'])
pop_median = np.median(input_data1['average_base_pay'])
print(pop_mean)
print(pop_median)
```

```python
plt.xlabel("High School Teacher Salaries by district")
plt.ylabel("School Districts")
plt.title("High School Teacher Salaries in Texas\n (sample)")
random_sample_1 = input_data1.take(np.random.permutation(len(input_data1))[:50])
plt.hist(random_sample_1.average_base_pay)
```

```python
np.mean(random_sample_1['average_base_pay'])
```

```python
#random_sample = input_data1.take(np.random.permutation(len(input_data1))[:25])
#sns.regplot(x=random_sample["District"], y=random_sample["average_base_pay"])
```

```python
stats.ttest_ind(input_data1.average_base_pay, random_sample_1.average_base_pay, equal_var = False)
```

```python
sns.distplot( input_data1["average_base_pay"] , color="skyblue", label="Public High School")
sns.distplot( input_data2["average_base_pay"] , color="red", label="Charter School").set_title("Mean Salaries of Public and Charter High Schools")
```

# Choose two continuous variables, and plot them three different ways.


```python
input_data3 = pd.read_csv("colrealestate.csv")
```

```python
graph1 = sns.lmplot(y="price", x="size", data=input_data3, fit_reg=False)
graph1.set_ylabels("Price in thousands")
graph1.set_xlabels("Size in square feet")
plt.title("2003 Colorado Home Prices")
```

```python
graph2 = sns.jointplot(y="price", x="size", data=input_data3, kind="reg")

plt.title("2003 Colorado Home Prices")
```

```python
plt.figure(figsize=(10,6))

graph3 = sns.violinplot(y="price", x="size", data=input_data3, kind="reg", inner=None)

sns.swarmplot(y="price", x="size", data=input_data3, color='k')

plt.title("2003 Colorado Home Prices")
```
